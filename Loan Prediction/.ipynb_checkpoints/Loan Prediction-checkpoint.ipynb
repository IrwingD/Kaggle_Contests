```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
import numpy as np
from sklearn import preprocessing
import xgboost
```

```python
df=pd.read_csv("train_u6lujuX_CVtuZ9i.csv")
```

```python
df.head()
```

```python
df['Gender'].value_counts()
```

```python
df[df['Gender'].isnull()]
```

```python
df[df['Dependents'].isnull()]
```

```python
df.Dependents.replace(['3+'],[3],inplace=True)
```

```python
df['Gender'].fillna('-999999999',inplace=True)
df['Married'].fillna('-999999999',inplace=True)
df['Education'].fillna('-999999999',inplace=True)
df['Self_Employed'].fillna('-999999999',inplace=True)
df['Property_Area'].fillna('-999999999',inplace=True)
df['Dependents'].fillna('0',inplace=True)
df.fillna(0,inplace=True)
```

```python
df['Dependents'].value_counts()
```

```python
df['Dependents']=df.Dependents.astype('float64')
df.dtypes
```

```python
df.drop('Loan_ID',1,inplace=True)
```

```python
df1=df.drop(['Dependents','ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term','Loan_Status'],1)
```

```python
df2=df[['Dependents','ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term','Loan_Status']]
```

```python
df.head()
```

```python
df1=df1.apply(LabelEncoder().fit_transform)
df=pd.concat([df1,df2],1)
```

```python
df.head()
```

```python
df.drop('Loan_Status',1).max()-df.drop('Loan_Status',1).min()
```

```python
scaler = preprocessing.MinMaxScaler()
scaled_df = scaler.fit_transform(df.drop('Loan_Status',1))
scaled_df = pd.DataFrame(scaled_df, columns=['Gender','Married','Education','Self_Employed',
                                             'Credit_History','Property_Area','Dependents','ApplicantIncome',
                                             'CoapplicantIncome','LoanAmount','Loan_Amount_Term'])
```

```python
scaled_df.max()-scaled_df.min()
```

```python
scaled_df.head()
```

```python
clf=LogisticRegression()
```

```python
X=np.array(scaled_df)
```

```python
y=np.array(df['Loan_Status'])
```

```python
clf.fit(X,y)
```

```python
clf.score(X,y)
```

```python
clf1=xgboost.XGBClassifier()
clf1.fit(X,y)
```

```python
clf1.score(X,y)
```

```python
df_test=pd.read_csv("test_Y3wMUE5_7gLdaTN.csv")
```

```python
df_test.Dependents.replace(['3+'],[3],inplace=True)
df_test['Gender'].fillna('-999999999',inplace=True)
df_test['Married'].fillna('-999999999',inplace=True)
df_test['Education'].fillna('-999999999',inplace=True)
df_test['Self_Employed'].fillna('-999999999',inplace=True)
df_test['Property_Area'].fillna('-999999999',inplace=True)
df_test['Dependents'].fillna('0',inplace=True)
df_test.fillna(0,inplace=True)
df_test['Dependents']=df_test.Dependents.astype('float64')


df_new=df_test.drop('Loan_ID',1)
df3=df_new.drop(['Dependents','ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term'],1)
df4=df_new[['Dependents','ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term']]
df3=df3.apply(LabelEncoder().fit_transform)
df_new=pd.concat([df3,df4],1)
```

```python
scaler = preprocessing.MinMaxScaler()
scaled_df_new = scaler.fit_transform(df_new)
scaled_df_new = pd.DataFrame(scaled_df_new, columns=['Gender','Married','Education','Self_Employed',
                                             'Credit_History','Property_Area','Dependents','ApplicantIncome',
                                             'CoapplicantIncome','LoanAmount','Loan_Amount_Term'])
```

```python
scaled_df_new.head()
```

```python
X=np.array(scaled_df_new)
X
```

```python
df_test['Loan_Status']=clf.predict(X)
```

```python
df_test['Loan_Status1']=clf1.predict(X)
```
